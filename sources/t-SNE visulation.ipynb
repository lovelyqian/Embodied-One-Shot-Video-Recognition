```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import manifold, datasets

plt.ion()

# load dataset
digits = datasets.load_digits(n_class=6)
X, y = digits.data, digits.target          

# X [n_samples,n_features]  y [label]



n_samples, n_features = X.shape
print('data:',X.shape, y.shape)

# to show the init data (digit datasets)
n = 20  # æ¯è¡Œ20ä¸ªæ•°å­—ï¼Œæ¯åˆ—20ä¸ªæ•°å­—
img = np.zeros((10 * n, 10 * n))
for i in range(n):
    ix = 10 * i + 1
    for j in range(n):
        iy = 10 * j + 1
        img[ix:ix + 8, iy:iy + 8] = X[i * n + j].reshape((8, 8))
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap=plt.cm.binary)
plt.xticks([])
plt.yticks([])
plt.show()



'''t-SNE'''
tsne = manifold.TSNE(n_components=2, init='pca', random_state=501)
X_tsne = tsne.fit_transform(X)

print("Org data dimension is {}ï¼ŒEmbedded data dimension is {}".format(X.shape[-1], X_tsne.shape[-1]))

'''åµŒå…¥ç©ºé—´å¯è§†åŒ–'''
x_min, x_max = X_tsne.min(0), X_tsne.max(0)
X_norm = (X_tsne - x_min) / (x_max - x_min)  # å½’ä¸€åŒ–
plt.figure(figsize=(8, 8))
for i in range(X_norm.shape[0]):
    plt.text(X_norm[i, 0], X_norm[i, 1], str(y[i]), color=plt.cm.Set1(y[i]), 
             fontdict={'weight': 'bold', 'size': 9})
plt.xticks([])
plt.yticks([])
plt.show()
```
